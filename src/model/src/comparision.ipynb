```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Calling the metrics of the different vaccination strategies.
%store -r random_strategy_metrics
%store -r articulation_points_strategy_metrics
%store -r degree_based_strategy_metrics
%store -r occupation_based_strategy_metrics
%store -r bmi_based_strategy_metrics
%store -r betweenness_based_strategy_metrics
%store -r eigenvector_based_strategy_metrics
```

```python
# A multibar plot to visualise the comparision of different metrics in vaccination strategies
X = np.arange(len(random_strategy_metrics))
f, ax = plt.subplots(figsize=(18,8))
ax.bar(X, random_strategy_metrics.values(), width=0.05, align='center')
ax.bar(X+0.05, articulation_points_strategy_metrics.values(), width=0.05, align='center')
ax.bar(X+0.1, degree_based_strategy_metrics.values(), width=0.05, align='center')
ax.bar(X+0.15, occupation_based_strategy_metrics.values(), width=0.05, align='center')
ax.bar(X+0.20, bmi_based_strategy_metrics.values(), width=0.05, align='center')
ax.bar(X+0.25, betweenness_based_strategy_metrics.values(), width=0.05, align='center')
ax.bar(X+0.30, eigenvector_based_strategy_metrics.values(), width=0.05, align='center')

plt.legend(('Random Strategy', 'Articulation Points Strategy', 'Degree Based Strategy', 
            'Occupation Based Strategy', 'BMI Based Strategy', 'Betweenness Based Strategy', 
            'Eigenvector Based Strategy'), loc='upper right')
plt.xticks(X+0.15, occupation_based_strategy_metrics.keys())
plt.title("Comparision between different strategies", fontsize=17)
plt.show()
```
