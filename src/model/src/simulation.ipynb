```python
from src.utils import SIR, run
```

```python
df_path = 'data/generation/contact_network/'
metadata_path = "data/generation/metadata/"
def get_path(path):
    retval = []
    for file in os.listdir(path):
        if file.endswith(".csv"):
            retval.append(os.path.join((path), file))
            req_path = max(retval)
            file = pd.read_csv(req_path)
    return file

df = get_path(df_path)
metadata = get_path(metadata_path)
```

```python
model = SIR(
    metadata=metadata,
    df=df
)
```
