```python
import os
import sys
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sys.path.append('../')
```

```python
from utils import SIARD, run
```

```python
df_path = 'data/generation/contact_network/'
metadata_path = 'data/generation/metadata/'
def get_path(path):
    retval = []
    for file in os.listdir(path):
        if file.endswith(".csv"):
            retval.append(os.path.join((path), file))
            req_path = max(retval)
            file = pd.read_csv(req_path)
    return file

df = get_path(df_path)
metadata = get_path(metadata_path)
```

```python
model = SIARD(
    metadata=metadata,
    df=df
)
```

## Random Strategy

```python
from vaccination_strategy.random_strategy import vaccination_strategy
vaccinated = vaccination_strategy(model)
run(model, vaccinated, 5)
```

```python
with open('epidemic_statistics.json') as file:
    data = json.load(file)
random_strategy_metrics = data['metrics']
%store random_strategy_metrics
os.remove('epidemic_statistics.json')
```
